### 가우시안 혼합 모델

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X,y= make_moons(n_samples=1000, noise=0.05)
```

```python
from sklearn.mixture import GaussianMixture

gm= GaussianMixture(n_components=3, n_init=10)
gm.fit(X)
```

```python
gm.weights_
```

```python
gm.means_
```

```python
gm.covariances_
```
